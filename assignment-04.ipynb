## Objective
Build a simple question-answering chatbot using a pre-trained transformer model.

```python
# Import libraries
from transformers import pipeline
```

## Load Pre-trained QA Model
We use a Hugging Face pipeline for question answering.

```python
# Load QA pipeline
qa_pipeline = pipeline("question-answering")
```

## Define Context
The chatbot answers questions based on this paragraph.

```python
context = """
Artificial Intelligence is a branch of computer science that aims to create intelligent machines.
It includes machine learning, natural language processing, and deep learning.
AI is widely used in healthcare, finance, and automation.
"""
```

## Ask Questions

```python
def ask_question(question):
    result = qa_pipeline(question=question, context=context)
    return result['answer']

# Example
question = "What is AI?"
print("Answer:", ask_question(question))
```

## Interactive Chatbot

```python
while True:
    q = input("Ask a question (type 'exit' to quit): ")
    if q.lower() == 'exit':
        break
    print("Answer:", ask_question(q))
```
