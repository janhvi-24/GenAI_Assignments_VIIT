## Objective
Implement and compare different prompt engineering techniques:
- Interview Approach
- Chain of Thought (CoT)
- Tree of Thought (ToT)
- Zero-shot Prompting
- Few-shot Prompting


## 1. Zero-shot Prompting

The model is given a task without examples.


```python
# Zero-shot example
prompt = "Classify the sentiment of this sentence: 'The movie was fantastic!'"
print(prompt)
```

## 2. Few-shot Prompting

The model is given a few examples before the actual query.


```python
# Few-shot example
prompt = '''
Sentence: I love this product!
Sentiment: Positive

Sentence: This is terrible.
Sentiment: Negative

Sentence: The experience was amazing!
Sentiment:
'''
print(prompt)
```

## 3. Chain of Thought (CoT)

Encourages step-by-step reasoning.


```python
# CoT example
prompt = "If a train travels 60 km in 1 hour, how far in 5 hours? Let's think step by step."
print(prompt)
```

## 4. Tree of Thought (ToT)

Explores multiple reasoning paths before selecting the best one.


```python
# ToT conceptual example
prompt = "Solve: What is the best strategy to minimize cost in a business? Explore multiple approaches before answering."
print(prompt)
```

## 5. Interview Approach

Simulates iterative questioning to refine answers.


```python
# Interview-style prompt
prompt = "Ask me questions one by one to understand my project requirements before giving a solution."
print(prompt)
```

## Comparison Summary

| Approach | Key Idea | Use Case |
|----------|--------|---------|
| Zero-shot | No examples | Simple tasks |
| Few-shot | Few examples | Pattern-based tasks |
| CoT | Step-by-step reasoning | Math, logic |
| ToT | Multiple reasoning paths | Complex problem solving |
| Interview | Iterative questioning | Requirement gathering |
