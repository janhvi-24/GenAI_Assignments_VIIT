```
from moviepy.editor import TextClip, CompositeVideoClip
from gtts import gTTS

# Text for animation
text = "Welcome to AI-based video animation"

# Generate speech
tts = gTTS(text)
tts.save("voice.mp3")

# Create text animation
txt_clip = TextClip(text, fontsize=50, color='white', size=(640, 480))
txt_clip = txt_clip.set_duration(5)

# Combine into video
video = CompositeVideoClip([txt_clip])
video = video.set_audio("voice.mp3")

# Export video
video.write_videofile("output.mp4", fps=24)

print("Video generated as output.mp4")
```
