```python
import pyttsx3

# Initialize the TTS engine
engine = pyttsx3.init()

# Input text
text = "Hello, this is a basic text to speech system using Tacotron."

# Save to file
engine.save_to_file(text, 'output.wav')
engine.runAndWait()

print("Audio generated and saved as output.wav")
```
