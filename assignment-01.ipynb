## Objective
Build a simple text generator using Markov Chains.


## Theory

A Markov Chain is a stochastic model that predicts the next state based only on the current state.

In text generation:
- Words are treated as states.
- The next word is chosen based on probabilities learned from training text.


```python
import random

def build_markov_chain(text):
    words = text.split()
    markov_chain = {}

    for i in range(len(words) - 1):
        word = words[i]
        next_word = words[i + 1]

        if word not in markov_chain:
            markov_chain[word] = []

        markov_chain[word].append(next_word)

    return markov_chain

def generate_text(chain, length=20):
    word = random.choice(list(chain.keys()))
    result = [word]

    for _ in range(length - 1):
        if word in chain:
            word = random.choice(chain[word])
            result.append(word)
        else:
            break

    return " ".join(result)

# Sample text
text = """Machine learning is fun and powerful. Learning machines can generate text using models."""

chain = build_markov_chain(text)
generated = generate_text(chain, 20)

print("Generated Text:")
print(generated)
```
